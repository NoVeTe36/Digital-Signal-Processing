```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

# Import Image and using OpenCV to remove noise

```python
noisy_image = cv2.imread('NoisyImage3.jpg')
dst = cv2.fastNlMeansDenoisingColored(noisy_image,None,11,6,7,21)
plt.imshow(noisy_image)
```

```python
row, col, ch = noisy_image.shape
fig, axs = plt.subplots(1, 2, figsize=(15, 15))
fig.tight_layout()
axs[0].imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[1].imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
axs[1].set_title('Fast Means Denoising')
plt.show()
```
